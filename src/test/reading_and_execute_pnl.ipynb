```python
import pandas as pd
from datasets import load_dataset
import spacy
from nltk.corpus import stopwords
import nltk

nltk.download('stopwords')

# carregar stop words e modelo de NLP
nlp = spacy.load('pt_core_news_sm')
stop_words = set(stopwords.words('portuguese'))

# carregar o dataset
dataset = load_dataset("ruanchaves/b2w-reviews01")
df = dataset['train'].to_pandas()

# definir as colunas 
columns_to_clean = ['review_title', 'reviewer_state', 'site_category_lv2', 'reviewer_gender', 'overall_rating', 'recommend_to_a_friend', 'reviewer_birth_year']
```

```python
import unicodedata
import re

def clean_text(text):
    # Converter para minúsculas
    text = text.lower()
    # Remover múltiplos espaços
    text = re.sub(r'\s+', ' ', text).strip()
    return text

# remove  ! , .
def remove_exclamations_and_periods(text):
    text = re.sub(r'[!.,]', '', text)
    return text

# remove stop words ("eu", "este", "e", "muito", "de")
def remove_stop_words(text):
    return ' '.join([word for word in text.split() if word not in stop_words])


# remove acentos e caracteres especiais
def remove_accents(text):
    text = unicodedata.normalize('NFD', text)
    text = re.sub(r'[\u0300-\u036f]', '', text)
    return text

```

```python
# aplicar as funções de limpeza
for column in columns_to_clean:
    df[column] = df[column].apply(lambda x: clean_text(str(x)))
    df[column] = df[column].apply(lambda x: remove_exclamations_and_periods(str(x)))
    df[column] = df[column].apply(lambda x: remove_accents(str(x)))
    df[column] = df[column].apply(lambda x: remove_stop_words(str(x)))
```
